### Datetime como tipo de dato

```python
df_meteorites['year']
```

Supongamos que queremos saber sobre un meteorito que cayó en cierta fecha, nosotros utilizaríamos la columna *year* pero como se contienen datos de tipo *str* solo podríamos hacer:

```python
df_meteorites[df_meteorites['year'] == '01/01/1976 12:00:00 AM']
```

Una desventaja que podemos notar es que sólo nos traera los registros que cumplen exactamente la condición, por lo que los registros que no tengan la misma hora no aparecen. Otra desventaja es que no podemos aplicar consultas con rangos de tiempo.

Es por eso que convertiremos nuestra columna *year* a tipo de dato datetime

```python
pd.to_datetime(df_meteorites['year'])
```

Al ejecutar la celda anterior nos marca un error, ya que necesitamos especificarle dos cosas:
    - Cómo es el formato de la fecha 
    - Qué hacer con los registros que no cumplen ese formato

Estos son los valores que puede tomar *errors*

errors{‘ignore’, ‘raise’, ‘coerce’}, default ‘raise’
    - If ‘raise’, then invalid parsing will raise an exception.

    - If ‘coerce’, then invalid parsing will be set as NaT.

    - If ‘ignore’, then invalid parsing will return the input.
    
Para establececer un formato:
 - **%m**: Mes
 - **%d**: Día
 - **%Y**: Año
 - **%H**: Hora
 - **%M**: Minuto
 - **%S**: Segundo
 - **%p**: AM o PM

Vemos que nuestra fecha tiene el siguiente formato: 

01/01/1976 12:00:00 AM

mes-dia-año hora-minutos-segundo am/pm

Por lo tanto sustituimos los valores por los definidos anteriormente

```python
pd.to_datetime(df_meteorites['year'], errors = 'coerce',format = '%m/%d/%Y %H:%M:%S %p')
```

Si ahora pedimos los tipos de datos que hay en nuestro Data Frame vemos que aún no se actualiza, tenemos que cambiar la columna *year* a lo que nos arrojó la celda pasada

```python
df_meteorites.dtypes
```

```python
df_meteorites['year'] =pd.to_datetime(df_meteorites['year'], errors = 'coerce',format = '%m/%d/%Y %H:%M:%S %p')
```

```python
df_meteorites.dtypes
```

Ahora si podemos aplicar filtros

```python
df_meteorites[df_meteorites['year'] >= pd.to_datetime('2000-01-01 00:00:00', format = '%Y-%d-%m %H:%M:%S')]
```

### Renombrar columnas

Podemos cambiar el nomre de columnas utilizando el método **rename()** y pasando como parámetro columns un diccionario que contenga como llaves los nombres actuales del Data Frame y como valores los nuevos nombres.

```python
df_meteorites.rename(columns = {'mass (g)':'mass'})
```

Podemos observar que si mostramos de nuevo el Data Frame no cambio el nombre

```python
df_meteorites
```

Para modificar atributos por medio de métodos de nuestro Data Frame podemos utilizar el parámetro **inplace**:

```python
df_meteorites.rename(columns = {'mass (g)':'mass'}, inplace = True)
```

```python
df_meteorites
```

### Eliminar columnas

Para borrar elementos de nuestro Data Frame utilizamos **drop()** ingresando como primer parámetro una lista de elementos a borrar. Puede ser una lista de indices o de nombres de columnas, para decirle qué estamos borrando necesitamos ocupar el parámetro **axis** donde 1 son columnas y 0 son filas. 

```python
df_meteorites.drop(['ones'], axis= 1, inplace = True)
```

```python
df_meteorites
```

### Copiando Data Frames

 Copiando un dataframe. Si nosotros igualamos una variable a un dataframe, los cambios que se hagan a ésta variable también las sufrirá el dataframe. Para evitar eso añadimos el parámetro **deep = True**


```python
df = df_meteorites.copy(deep = True)
df
```
